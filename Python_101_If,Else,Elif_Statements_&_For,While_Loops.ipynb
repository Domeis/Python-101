## If/Else/Elif statements 
Programming often involves examining a set of conditions and deciding which action to take based on those conditions. Python's <b>*if*</b> statements allows you to examine the current state of a program and respond appropriately to that state.
 
 - At the heart of every <b>*if*</b> statement is an expression that can be evaluated as TRUE or FALSE and is called a conditional test. It's used to decide whether the code in an if statement should be executed.

The general syntax for a simple if statement is 
```` python 
if condition: 
    do_something
elif condition2:
    another_action
elif condition3:
    another_action
else condition4:  
    another_action
````
If the condtion is true, then do the indented statements. <br> If the condition is not true, then skip the indented statements. 

- We can set a condition where if the "if" condition is not satifised(FALSE), we follow another set of instructions to follow through the "elif" and "else" statements.

- There can be <b> multiple "elif" statements </b> but there can be <b> only one "else."</b>(although it's not mandatory)
- Else statement can be seen as a "catch all" condition. 

`````python
age = 101
if age < 19:
    print("You are not old enough")
elif age >= 100:
    print("You are too old")
else:
    print("You are old enough ")
`````

## For and While loops
- *for* loops are traditionally used when you have a block of code you want to repeat a fixed number of times. <br>



```` python
computer_brands = ["Apple", "Asus", "Dell", "Samsung"]
for brands in computer_brands: 
    print(brands)
    
````
This basically says: For <b> all elements </b> in <b> a list </b>, do the following. <br>

 > More specifically, for every element that we assign the variable brands in the list computer_brands, print out the variable brands.

- The *while* loops are used when you want to run a code as long as a certain condition is met. It will keep repeating itself until the condiion is false.  



```` python
i = 1
while i < 6:
  print(i)
  i += 1
````

This says: While this is true, do this.
> More specifically, as long as the value of the variable "i" is less than 6, print i and add 1 to i each time the loop is executed. 

`````python
computer_brands = ["Apple", "Asus", "Dell", "Samsung"]
for brands in computer_brands: 
    print(brands)
`````

`````python
i = 1
while i < 6:
  print(i)
  i += 1
`````
